```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pilih fitur numerik yang sensitif terhadap outlier
numeric_cols = [
    'Age', 'MonthlyIncome', 'DistanceFromHome', 'TotalWorkingYears',
    'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager'
]

# Buat boxplot untuk masing-masing fitur
plt.figure(figsize=(14, 12))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```


MonthlyIncome: Ada nilai sangat tinggi (right-skewed).

DistanceFromHome: Beberapa outlier, terutama di atas 20 km.

TotalWorkingYears, YearsAtCompany, YearsWithCurrManager: Ada lonjakan ekstrem di nilai maksimal.

YearsSinceLastPromotion: Outlier terlihat jelas di atas 10 tahun.

Interpretasi:
Outlier ini masih bermakna secara kontekstual (misalnya: orang yang sudah kerja 40 tahun bisa saja ada), jadi tidak harus dibuang, tapi bisa distandarisasi atau log-transform jika model sensitif terhadap skala atau distribusi.
